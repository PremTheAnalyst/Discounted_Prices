```python
df_cat = pd.get_dummies(df['maincateg'])
df = pd.concat([df, df_cat], axis=1)
df_1 = pd.get_dummies(df['platform'])
df = pd.concat([df, df_1], axis=1)
df.head()
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df1 = le.fit_transform(df.title)
df1
```

```python
df2 = pd.DataFrame(df1, columns =['new_title'])
df2.head()
```

```python
df = pd.concat([df, df2], axis = 1)
df.head()
```

```python
df = df.drop(['maincateg','platform' , 'title', 'Offer %'], axis = 1)
```

```python
df.isnull().sum()
```

```python
df = df.fillna(df.mean())
```

```python
x = df.drop('price1', axis = 1)
y = df.price1
```

```python
x.shape
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2)
```

```python
from sklearn.preprocessing import StandardScaler
s1 = StandardScaler()
x_train_scaled = s1.fit_transform(x_train, y_train)
y_test_scaled = s1.fit_transform(x_test, y_test)
x_train_scaled
```

```python
from sklearn.linear_model import LinearRegression
m1 = LinearRegression()
```

```python
m1.fit(x_train_scaled, y_train)
```

```python
m1.score(y_test_scaled, y_test)
```

```python
m1.predict(y_test_scaled)
```

```python
ndf8.isnull().sum()
```
